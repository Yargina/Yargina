```python
from pylab import rcParams
rcParams['figure.figsize'] = 10,7
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cacao = pd.read_csv('cacao.csv')
cacao.head(100)
```

```python
cacao_dynamics = cacao.groupby(['Company\nLocation']).sum().sort_values(by='Rating', ascending=False).head(10)
cacao_dynamics.plot(kind='pie', y='Rating')
```

# В Соединенных штатах рейтинг шоколада наивысший. Стоит обратить на него внимание. 

```python
cacao_barh = cacao[cacao['Company\nLocation']=='U.S.A.'].sort_values('Rating', ascending=False).head(15)
cacao_barh
```

# Рейтинг 4.0 самый высокий в США.

```python
plt.barh(cacao_barh['Specific Bean Origin\nor Bar Name'], cacao_barh['REF'])
```

# Наименее обозреваемые какао-бобы Мадагаскарские. Цена и популяция плодоносящих деревьев. 

```python
plt.barh(cacao_barh['Specific Bean Origin\nor Bar Name'], cacao_barh['REF'])
plt.xlabel('Количество REF')
plt.ylabel('Разновидность')
plt.title('Топ 15 разновидностей какао-бобов по обзорам')
plt.text(2100, -0.1, 'Наименее обозреваемый')
plt.grid(15)
```

# Наиболее обозреваемые какао-бобы Phantom. Возможно, неприхотливы к климатическим условиям.

```python
plt.barh(cacao_barh['Cocoa\nPercent'], cacao_barh['REF'])
plt.xlabel('Количество REF')
plt.ylabel('Процентное содержание какао')
plt.title('Топ 6 разновидностей процентного содержания какао по обзорам')
plt.text(2100, -0.1, 'Наиболее обозреваемый')
plt.grid(6)
```

# Наиболее обозреваемый шоколад с содержанием какао 70 %. На втором месте 74 %, что обусловлено горечью шоколада и полезностью. 

```python
cacao_d = cacao.groupby('Rating').mean()
```

```python
cacao_d.plot()
```

# Наиболее обозреваемые марки шоколада с рейтингом от 2,3 до 3,8.

```python
cacao_dynamics = cacao.groupby([cacao.Company\nLocation,'Company\nLocation']).sum()
cacao_dynamics = cacao_dynamics.loc[(cacao_dynamics.index.get_level_values(0).isin(['Belgium', 'Switzerland']))]
cacao_dynamics.unstack().plot(kind='bar')
```

```python
cacao_hist = cacao.Rating.hist()
cacao_hist.set_title('Гистограмма распределения рейтинга по обзорам')
cacao_hist.set_xlabel('Оценка')
cacao_hist.set_ylabel('Количество обзоров')
```
